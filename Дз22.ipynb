```python
import sympy as sym
x = sym.Symbol('x')
sym.diff(x**5+sym.sin(2*x)+sym.cos(3*x+3)).subs(x,1)
```

```python
from sympy.plotting import plot
plot((sym.sin(2*x+1))**5, (x, -5, 5), title = 'График функции y = (sin(2x+1))^5')
```

```python
plot(sym.diff((sym.sin(2*x+1))**5), (x, -5, 5), title = 'График производной функции y = (sin(2x+1))^5')
```

```python
plot(sym.diff(sym.diff((sym.sin(2*x+1))**5)), (x, -5, 5), title = 'График второй производной функции y = (sin(2x+1))^5')
```
