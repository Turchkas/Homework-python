```python
import pandas as pd
import requests
import time
from bs4 import BeautifulSoup
```

```python
def habr(query, numbers):
    df = pd.DataFrame()
    URL = f'https://habr.com/ru/search/page{numbers}'

    for q in query:
        params = {'q' : q, 'target_type': 'posts', 'order': 'relevance'}
        req = requests.get(URL, params = params)
        time.sleep(0.3)
        soup = BeautifulSoup(req.text)
        articles = set(soup.find_all('article', class_ = 'tm-articles-list__item'))
        for article in articles:
            try:
                title = article.find('h2', class_ = "tm-title tm-title_h2").text
                link = 'https://habr.com' + article.find('h2', class_ = "tm-title tm-title_h2").find('a').get('href')
                date = article.find('span', class_ = "tm-article-datetime-published").text.strip()
                like = article.find('span', class_ = "tm-icon-counter__value").text
                row = {'date': date,' title': title, 'link': link, 'likes': like}
                df = pd.concat([df, pd.DataFrame([row])])
            except:
                title = 'Кривой заголовок'
                link = 'Кривая ссылка'
                date = 'Кривая дата'
                like = 'Кривые данные'
                row = {'date': date,' title': title, 'link': link, 'likes': like}
                df = pd.concat([df, pd.DataFrame([row])])
    return df.reset_index(drop = True)
result = habr(['python', 'анализ данных'], 1)
result
```

```python
def add_full_text(posts_df):
    i = 0
    for el in posts_df['link']:
        try:
            req = requests.get(el).text
            soup = BeautifulSoup(req)
            time.sleep(0.3)
            full_text = soup.find('div', class_='tm-article-body').text.strip()
            posts_df.loc[i, 'text'] = full_text
            i += 1
        except:
            full_text = 'Кривой текст'
            posts_df.loc[i, 'text'] = full_text
            i += 1
    return posts_df.reset_index(drop = True)

add_full_text(result)

#Почему, если выводить posts_df.reset_index(drop = True).reindex(columns = ['date', 'title', 'link', 'text', 'likes']), то столбец title заполняется весь NaN?
```
