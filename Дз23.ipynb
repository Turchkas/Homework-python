```python
#Задание 3
import numpy as np
import matplotlib.pyplot as plt

fig = plt.figure()
X = np.arange(-10,10,0.1)
Y = np.arange(-10,10,0.1)
X,Y = np.meshgrid(X,Y)

Z = Y**2 - X**2*Y + 2*X*Y

axes = fig.add_subplot(projection='3d')
axes.plot_surface(X, Y, Z)
```
