```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import IsolationForest
from sklearn.metrics import accuracy_score
from scipy.stats import shapiro
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('glass.csv')
df.info()
```

```python
df
```

```python
X_train, X_test, y_train, y_test = train_test_split(df.loc[:, ~df.columns.isin(['Type'])], df['Type'], random_state=42)
# Обучение модели случайного леса
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)
# смотрим разницу факта и прогноза для случайного леса
result_clf_rf = pd.DataFrame([np.array(y_test), clf_rf.predict(X_test)]).T
result_clf_rf.columns = ['y_test', 'y_pred']
result_clf_rf.head()
```

```python
# Оценка точности предсказания модели
accuracy = accuracy_score(y_test, clf_rf.predict(X_test))
print(f'Точность предсказания модели: {accuracy}')
```

```python
# Визуализация распределения значений каждой переменной
for column in df.columns:
    if column != 'Type':
        plt.figure(figsize=(8, 6))
        sns.histplot(df[column], kde=True)
        plt.xlabel('Значение')
        plt.ylabel('Количество')
        plt.title(f'Распределение столбца {column}')
        plt.show()
        stat, p = shapiro(df['Na'])
        print(f'Тест Шапиро-Уилка для колонки {column}: p-value={p}')
        if p > 0.05:
            print(f'Столбец {column} имеет нормальное распределение')
        else:
            print(f'Столбец {column} имеет не нормальное распределение')
```

```python
for column in df.columns:
    if column != 'Type':
        plt.figure(figsize=(8, 6))
        plt.boxplot(df[column], vert=True)
        plt.title(f'Boxplot столбца {column}')
        plt.ylabel(f'Значение {column}')
        plt.show()
        print(f"Диапазон значений столбца {column} - ({np.percentile(df[column], 25)-1.5*(np.percentile(df[column], 75)-np.percentile(df[column], 25))}, {np.percentile(df[column], 75)+1.5*(np.percentile(df[column], 75)-np.percentile(df[column], 25))})")
```

```python
#Удалим строки, значения которые выходят из целевого диапазона
df_temp = df.copy()
for column in df_temp.columns:
    if column != 'Type':
        df_temp[column] = df_temp[column].loc[(df_temp[column]>=np.percentile(df_temp[column], 25)-1.5*(np.percentile(df_temp[column], 75)-np.percentile(df_temp[column], 25))) & (df_temp[column]<=(np.percentile(df_temp[column], 75)+1.5*(np.percentile(df_temp[column], 75)-np.percentile(df_temp[column], 25))))]
df_temp = df_temp.dropna()
df_temp.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df_temp.loc[:, ~df_temp.columns.isin(['Type'])], df_temp['Type'], random_state=42)
# Обучение модели случайного леса
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)
# смотрим разницу факта и прогноза для случайного леса
result_clf_rf = pd.DataFrame([np.array(y_test), clf_rf.predict(X_test)]).T
result_clf_rf.columns = ['y_test', 'y_pred']
result_clf_rf.head()
```

```python
# Оценка точности предсказания модели
accuracy = accuracy_score(y_test, clf_rf.predict(X_test))
print(f'Точность предсказания модели: {accuracy}')
```

```python
df1 = df.copy()
df1.shape
```

```python
#С помощью изолирующего леса удалим выбросы
iso = IsolationForest()
y_outliers = iso.fit_predict(df1)
#отбрасывание строк с выбросами
for i in range(len(y_outliers)):
    if y_outliers[i] == -1:
        df1.drop(i, inplace = True)
#проверка новой формы датафрейма
df1.shape
```

```python
X_train, X_test, y_train, y_test = train_test_split(df1.loc[:, ~df1.columns.isin(['Type'])], df1['Type'], random_state=42)
# Обучение модели случайного леса
clf_rf = RandomForestClassifier()
clf_rf.fit(X_train, y_train)
# смотрим разницу факта и прогноза для случайного леса
result_clf_rf = pd.DataFrame([np.array(y_test), clf_rf.predict(X_test)]).T
result_clf_rf.columns = ['y_test', 'y_pred']
result_clf_rf.head()
```

```python
# Оценка точности предсказания модели
accuracy = accuracy_score(y_test, clf_rf.predict(X_test))
print(f'Точность предсказания модели: {accuracy}')
```
