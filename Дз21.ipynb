```python
#Задание 1
import numpy as np
import matplotlib.pyplot as plt

A = [
[-100, -100, 1],
[-100, 100, 1],
[100, 100, 1],
[0, 0, 1],
[100, -100, 1],
[-100, -100, 1]
]

A = np.array(A)

x = A[:,0]
y = A[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

```python
B = np.array(A) @ np.array([[1/2,0,0], [0,1/2,0], [0,0,1/2]])
x = B[:,0]
y = B[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

```python
C = np.array(A) @ np.array([[np.cos(130*np.pi/180), np.sin(130*np.pi/180), 0], [-(np.sin(130*np.pi/180)), np.cos(130*np.pi/180), 0], [0, 0,1]])

x = C[:,0]
y = C[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```

```python
D = np.array(A) @ np.array([[0, 1, 0], [1, 0, 0], [0, 0, 1]])

x = D[:,0]
y = D[:,1]
plt.plot(x, y)
plt.ylim([-200, 200])
plt.xlim([-200, 200]) 
```
