Так как rectal temperature коррелирует с temperature of extremities, то заменим пропуски следующим образом: где temperature of extremities = 1, 3 или 4, то rectal temperature = 37.8, где temperature of extremities = 2, то ставим медианную температуру. Где оба значения не заполнены - удалим их

```python
horse = horse.loc[~((horse['rectal temperature'].isna()) & (horse['temperature of extremities'].isna()))]
horse
```

```python
horse.loc[((horse['rectal temperature'].isna()) & (horse['temperature of extremities'] != 2)), 'rectal temperature'] = 37.8
```

```python
horse.loc[((horse['rectal temperature'].isna()) & (horse['temperature of extremities'] == 2)), 'rectal temperature'] = horses['rectal temperature'].median()
```

Пульс связан с частотой дыхания. Сделаем замену следующим образом: где оба значения не заполнены - удалим их, где частота дыхания больше 10, ставим медианный пульс, где меньше 10 - 30

```python
horse = horse.loc[~((horse['pulse'].isna()) & (horse['respiratory rate'].isna()))]
horse
```

```python
horse.loc[((horse['pulse'].isna()) & (horse['respiratory rate'] > 10)), 'pulse'] = horses['pulse'].median()
```

```python
horse.loc[((horse['pulse'].isna()) & (horse['respiratory rate'] <= 10)), 'pulse'] = 30
```

```python
horse[horse['respiratory rate'].isna()]
```

Видим, что у всех незаполненных значений respiratory rate пульс выше нормы, заменим на медианное значение

```python
horse.loc[horse['respiratory rate'].isna(), 'respiratory rate'] = horses['respiratory rate'].median()
```

```python
horse[horse['temperature of extremities'].isna()]
```

Так как rectal temperature коррелирует с temperature of extremities, то заменим пропуски следующим образом: где rectal temperature от 37.6 до 38.0, то temperature of extremities = 1, если rectal temperature больше 38,0 то temperature of extremities = 2, если rectal temperature от 37.2 до 37.6, то temperature of extremities = 3, в остальном случае ставим 4

```python
horse.loc[horse['rectal temperature'] > 38.0, 'temperature of extremities'] = 2
```

```python
horse.loc[(horse['rectal temperature'] >= 37.6) & (horse['rectal temperature'] <= 38.0), 'temperature of extremities'] = 1
```

```python
horse.loc[(horse['rectal temperature'] < 37.6) & (horse['rectal temperature'] >= 37.2), 'temperature of extremities'] = 3
```

```python
horse.loc[horse['rectal temperature'] < 37.2, 'temperature of extremities'] = 4
```

Так как уровень боли достаточно субъективен для оценки, заменим на моду

```python
horse.loc[horse['pain'].isna(), 'pain'] = horses['pain'].mode()[0]
```
