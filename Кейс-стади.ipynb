```python
import pandas as pd
import numpy as np
import scipy.stats as stats
```

```python
#Задание 1
#Используем p=0.05 как пороговое значение
games = pd.read_csv('vgsales.csv')
games
```

```python
games_sports = games[games['Genre'] == 'Sports']
games_sports
```

```python
other_games = games[games['Genre'] != 'Sports']
other_games
```

```python
other_games['Critic_Score'].hist()
```

```python
games_sports['Critic_Score'].hist()
```

Нулевая гипотеза: Критики относятся к спортивным играм также, как и к играм другого жанра


Альтернативная гипотеза: Критики относятся к спортивным играм иначе, нежели к играм другого жанра

```python
stat, p = stats.ttest_ind(games_sports.dropna()['Critic_Score'], other_games.dropna()['Critic_Score'])

if p > 0.05:
    print("Не отклоняем нулевую гипотезу, различия статистически не значимы")
else:
    print("Отклоняем нулевую гипотезу, различия статистически значимы")
```

```python
games_sports.boxplot(column='Critic_Score')
```

```python
other_games.boxplot(column='Critic_Score')
```

Вывод: К спортивным играм критики относятся более благосклонно, чем к играм другого жанра

```python
#Задание 2
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
#nltk.download('stopwords')
stopwords_set = set(stopwords.words('english'))
#nltk.download('omw-1.4')
#nltk.download('wordnet')
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
sms = pd.read_csv('https://raw.githubusercontent.com/obulygin/pyda_homeworks/master/stat_case_study/spam.csv')
#Удаляем мусорные символы и переводим к нижнему регистру
sms['Message'] = [re.sub('[\W_]+', ' ', text.lower()) for text in sms['Message']]
#Выкидываем стоп-слова
sms['Message'] = sms['Message'].apply(lambda x: ' '.join([word for word in x.split() if word not in stopwords_set]))
#Приводим слова к нормальной форме
sms['Message'] = sms['Message'].str.split().apply(lambda x: [wordnet_lemmatizer.lemmatize(word) for word in x]).apply(lambda x: ' '.join([word for word in x]))
sms
```

```python
#Преобразуем все сообщения в вектора TF-IDF
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(sms.Message)
names = tfidf.get_feature_names_out()
tfidf_matrix = pd.DataFrame(tfidf_matrix.toarray(), columns=names)
tfidf_matrix
```

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
y = le.fit_transform(sms.Category)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(tfidf_matrix, y, test_size=0.3, random_state=42)
model = LogisticRegression(random_state=42)
model.fit(X_train, y_train)
```

```python
# Рассчитаем предсказанные значения
y_pred = model.predict(X_test)
y_pred
```

```python
#Оценим точность на тестовых данных
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
y_all_pred = model.predict(tfidf_matrix)
y_all_pred
```

```python
#Опишем результаты при помощи confusion_matrix
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y, y_all_pred)
confusion_matrix
```
